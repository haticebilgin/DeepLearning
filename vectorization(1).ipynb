```
import numpy as np 

a= np.array([1,2,3,4])
print(a)
```

```
import time 

a= np.random.rand(1000000);
b= np.random.rand(1000000);

#Vectorization
tic = time.time()
c= np.dot(a,b)
toc = time.time()

print("Vectorized version: " + str(1000*(toc-tic))+"ms")
print(c)
#Non-Vectorization
c=0
i=0
tic = time.time()

for i in range(1000000):
  c += a[i]*b[i]

toc = time.time()

print("Non-Vectorization version: " + str(1000*(toc-tic))+"ms")
print(c)
```

```
#Let's say you already have a vector, v, in memory and you want to apply the exponential operation on every element of this vector v. 
#Let's call output vector u
import math
v= np.ones((10,1))

#Non-Vectorization
u= np.zeros((10,1))

tic = time.time()

for i in range(10):
  u[i]= math.exp(v[i])

toc = time.time()
print("Non-Vectorization version: " + str(1000*(toc-tic))+"ms")

#Vectorization

tic = time.time()
u= np.exp(v)
toc = time.time()
print("Vectorization version: " + str(1000*(toc-tic))+"ms")
```
