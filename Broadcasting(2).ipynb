```
# Calories from Carbs, Proteins, Fats in 100g of different foods are given.
# Calculate % of calories from Carbs, Proteins, Fats
```

```
import numpy as np

A=np.array([[56,0,4.4,68],[1.2,104,52,8],[1.8, 135, 99,0.9]])

print(A)
```

```
cal = A.sum(axis=0) #axis=0 sum of horizantal elements, axis=1 sum of vertical elements
print(cal)
```

```
percentage= 100*(A/cal)
print(percentage)
```
